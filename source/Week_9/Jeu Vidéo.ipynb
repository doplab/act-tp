## Fonctionnement

Un jeu vidéo est en fait un programme qui affiche de la donnée à l'écran toutes les 33 millisecondes à peu près. Entre chaque images, le jeu doit recalculer et mettre à jour la donnée pour le prochain affichage.


Dans la cellule suivante, vous avez à votre disposition la fonction `collision(data)`.

Le jeu appelle tout seul la fonction `collision` avec la donnée qu'il possède, vous pouvez ensuite modifier cette donnée, la mettre à jour et la rendre dans le return.

La donnée est une liste de dictionnaires. Chaque dictionnaire contient les clefs suivantes:

    {
        "x": 10 # La position sur l'axe x de l'angle du haut à gauche du personnage
        "y": 100 # La position sur l'axe y de l'angle du haut à gauche du personnage
        "spriteX": 20 # Influence quel personnage est affiché
        "spriteY": 20 # Influence quel personnage est affiché
        "angle": 2 * PI # L'angle vers lequel le personnage se dirige
        "deleted": False # Détermine si l'élément est effacé ou non
    }

**Veuillez noter que `print()` ne marchera pas dans cet exercice.**

Vous pouvez vous amuser à changer les valeurs des dictionnaires dynamiquement.

Pour changer de personnage, vous pouvez modifier les valeurs `spriteX` et `spriteY` par des multiples de 21.

Par exemple, pour n'avoir que des `Marios`, faites:

    def collision(data):
        for i in data:
            i["spriteX"] = 0
            i["spriteY"] = 0
        return data

Pour des `goombas volants`, faites:

    def collision(data):
        for i in data:
            i["spriteX"] = 166
            i["spriteY"] = 21
        return data

Pour que tous les personnages tournent en rond:

    def collision(data):
        for i in data:
            i["angle"] += 3.14 / 2
        return data

```python
from binder import Binder, start
from abc import ABC, abstractmethod

class AbstractCell(ABC):
    def __init__(self, alive, matrix = None):
        if matrix is not None:
            self.matrix = matrix
            self.up = matrix[0]
            self.left = self.matrix[1][0]
            self.right = self.matrix[1][2]
            self.down = self.matrix[2]
        self.alive = bool(alive)
        
    @abstractmethod
    def step(self):
        pass
    
    def __repr__(self):
        return str(int(self.alive))
    
    
class Cell(AbstractCell):
    def step(self):
        neighbours_count = 0
        for i in self.matrix:
            for j in i:
                if j.alive and j is not self:
                    neighbours_count += 1
        if neighbours_count >= 4:
            self.alive = False
        elif self.alive == False and neighbours_count >= 2 and neighbours_count <= 4:
            self.alive = True
        else:
            self.alive = False
        
class Game:    
    def step(self, data):
        copy = [i for i in data]
        length = len(data)
        for i in range(length):
            for j in range(length, 0, -1):
                matrix = [[Cell((copy[l] & (1 << k)) if l < length else 0) for k in range(j + 1, j - 2, -1)] for l in range(i - 1, i + 2)]
                cell = Cell(copy[i] & (1 << j), matrix=matrix)
                matrix[1][1] = cell
                prev_state = cell.alive
                cell.step()
                if prev_state != cell.alive:
                    if cell.alive == True:
                        data[i] += 1 << j
                    else:
                        data[i] = data[i] - (1 << j)
        return data
                
    def start(self):
        start(self.step)

# FIN DE VOTRE CODE
game = Game()
game.start()
```

```python
from binder import Binder, start
from abc import ABC, abstractmethod
from copy import deepcopy

        
class AbstractCell(ABC):
    def __init__(self, ptr, matrix = None, mutable = False):
        if matrix is not None:
            self.matrix = matrix
            self.up = matrix[0]
            self.left = self.matrix[1][0]
            self.right = self.matrix[1][2]
            self.down = self.matrix[2]
        self.mutable = mutable
        self.ptr = ptr
        self.disabled = bool(ptr["disabled"])
        
    def disable(self):
        if self.mutable:
            self.ptr["disabled"] = True
        
    def enable(self):
        if self.mutable:
            self.ptr["disabled"] = False
        
    @abstractmethod
    def step(self):
        pass
    
    
class Cell(AbstractCell):
    def step(self):
        if not self.left.disabled:
            self.enable()
        else:
            self.disable()
        
class Game:    
    def step(self, data):
        copy = deepcopy(data)
        length = len(data)
        for i in range(1, length-2):
            for j in range(1, length-2):
                matrix = [[Cell(copy[l][k]) for k in range(j - 1, j + 2)] for l in range(i - 1, i + 2)]
                cell = Cell(data[i][j], matrix=matrix, mutable = True)
                cell.step()
        return data
                
    def start(self):
        start(self.step)

# FIN DE VOTRE CODE
game = Game()
game.start()
```

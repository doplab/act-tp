# Complexité et Big O notation 

#### Dans ce document nous allons revoir les notions de complexité des algorithmes et en particulier la notion de Big O

La notation Big O à comme but d’exprimer la complexité temporelle des algorithmes dans le pire scenario d’exécution (Worst Case scenario) et suit quelques principes de base:

+ Les constantes sont ignorés

    $O(2n) = 2*O(n) = O(n)$ 


+ Les termes dominés sont ignorés 
    
    $O(2n^2+5n+50)$ = $O(n^2)$
    
    
+ Utilisez différentes variables pour différents inputs


+ **Uniquement le pire cas d'exécution est consideré** (a.e. en cas de test logique)

    
### Complexités fréquentes 

Les complexités fréquentes sont les suivantes: 

#### Constante $O(1)$

Cela veut dire que la complexité ne depend pas de la taille du vecteur en input. 
+ les méthodes append et pop sur les listes en python

#### Logarithmique $O(log(n))$ 

+ l’algorithme de recherche binaire 

#### Linaire $O(n)$

+ les boucles for et while

#### Quadratique, Cubique, etc… $O(n^2)$ , $O(n^3)$

+ des boucles imbriquées ayant le même nombre d’itération

+ l’algorithme de tri par selection à une complexité de $\frac{1}{2}(n^2-n) = O(n^2)$

#### Exponentielle $O(2^n)$

+ Fibonacci par recursion comme vu en exercice
